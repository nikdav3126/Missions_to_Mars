# Part One: Webscraping

### Section 1: Initial Scrape of Mars Webpage

```python
#import dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless = False)
```

```python
#visit the mars news site
url = "https://redplanetscience.com"
browser.visit(url)
#per Dr. A add wait time so that full page may load
browser.is_element_present_by_css('div.list_text', wait_time =1)
```

```python
#pull out html to do a search 
html= browser.html
news_soup = soup(html, "html.parser")
print(news_soup.prettify())
```

```python
#scrape for latest title
latest = news_soup.select_one("div.list_text")
latest_title = latest.find("div", class_= "content_title").get_text()
latest_title
```

```python
#scrape for latest paragraph
latest_p = latest.find("div", class_= "article_teaser_body").get_text()
latest_p
```

### Part 2: JPL Space Image

```python
#go to website for space images
url="https://spaceimages-mars.com/"
browser.visit(url)
```

```python
#looking for image based on button tag
featured_img = browser.find_by_tag("button")[1]
featured_img.click()
```

```python
#pull out gtml to parse through
html= browser.html
image_soup = soup(html, "html.parser")
print(image_soup.prettify())
```

```python
#find location of image
featured_img_find = image_soup.find("img", class_= "fancybox-image").get("src")
featured_img_find
```

```python
#add image location onto main website
featured_img_url = f'https://spaceimages-mars.com/{featured_img_find}'
featured_img_url
```

### Part 4: Mars Hemispheres

```python
#visit site
url = "https://marshemispheres.com/"
browser.visit(url)
```

```python
#create a list
hemisphere_image_urls = []

links = browser.find_by_css("a.product-item img")

#create for loop to parse through all links on url page
for i in range (len(links)):

    hemi_url = {}
    
    browser.find_by_css("a.product-item img")[i].click()
    
    sample = browser.links.find_by_text("Sample").first
    
    hemi_url["title"]=browser.find_by_css("h2.title").text
    
    hemi_url["img_url"] = sample["href"]
    
    hemisphere_image_urls.append(hemi_url)
    
    browser.back()
```

```python
hemisphere_image_urls
```

```python
browser.quit()
```
